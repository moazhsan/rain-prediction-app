```python
print(f'duplicated :{data.duplicated().sum()}')
print('========================')
print(f'isna : \n{data.isna().sum()}')
```

```python
for i in data.select_dtypes(include='number').columns:
    data[i].fillna(data[i].median(), inplace=True)

for i in data.select_dtypes(include=['object', 'category']).columns:
    if i != "RainTomorrow":   
        data[i].fillna(data[i].mode()[0], inplace=True)
data.isna().sum()
```

```python
data['RainTomorrow'].value_counts()
```

```python
data = data.dropna(subset=['RainTomorrow'])
```

```python
data.info()
```

```python
data=data.drop('Date',axis=1)
```

```python
cat = [i for i in data.columns if data[i].dtype=='O']
for i in cat:
    print(data[i].value_counts())
```

```python
cat
```

encoding

```python
for i in cat:
    data=pd.get_dummies(data, columns=[i], drop_first=True,dtype=int)
print(data.shape)
data.head(5)
```

split , Scaleing

```python
x = data.drop('RainTomorrow_Yes', axis=1)
y = data['RainTomorrow_Yes']
x
```

```python
joblib.dump(x.columns.tolist(), "features.pkl")
```

Scaleing

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x)
x = pd.DataFrame(x_scaled, columns=x.columns)
joblib.dump(scaler, "scaler.pkl")
```

```python


from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
print(type(x_train), x_train.shape)
print(type(y_train), y_train.shape)
print(type(x_test), x_test.shape)
print(type(y_test), y_test.shape)
y_train = y_train.values
y_test = y_test.values
```

model building

```python
import tensorflow as tf
import keras as ka
ka_model=ka.models.Sequential([
    ka.layers.Dense(8,activation='tanh'),
    ka.layers.Dense(128,activation='tanh'),
    ka.layers.Dense(256,activation='tanh'),
    # ka.layers.Dense(64,activation='tanh'),
    # ka.layers.Dense(16,activation='tanh'),
    ka.layers.Dense(8,activation='tanh'),

    ka.layers.Dense(1,activation='sigmoid')
])


MyOptimizer = tf.keras.optimizers.AdamW(
    learning_rate=0.001,
    weight_decay=0.004,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
    name="AdamW"
)

ka_model.compile(optimizer=MyOptimizer, loss='binary_crossentropy', metrics=['accuracy'])
#              loss,metrics
# regrssion = mes,mes
# BC = binary_crossentropy,accuracy
# MC = categorical_crossentropy,accuracy

history = ka_model.fit(
    x_train, y_train,
    validation_data=(x_test, y_test),
    epochs=100,
    batch_size=128,
    verbose=1,
    callbacks=[tf.keras.callbacks.EarlyStopping(
        patience=5,
        monitor='val_accuracy',
        restore_best_weights=True
    )]
)
```

```python
import matplotlib.pyplot as plt


plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Accuracy over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns

y_pred = (ka_model.predict(x_test) > 0.5).astype(int)

print(classification_report(y_test, y_pred))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='Blues')
plt.show()
```
